```python
import re
import random as rm
import pandas as pd
from urllib import request
holdb=[] #holding books 
              #      "https://www.gutenberg.org/cache/epub/68043/pg68043.txt"   worked url u could use to test the coode!
            #        "https://www.gutenberg.org/cache/epub/68039/pg68039.txt"    worked url u could use to test the coode!
book=input("Please enter books number")
url=[]
for i in range(0,int(book)):
    S=input("please enter the book URL ")
    url.append(S)
def getbook(url):
    response = request.urlopen(url)
    #print(response)
    raw = response.read()
    text = raw.decode("utf-8-sig")
    holdb.append(text)
for i in url:         #getting books and append them into the holdb list
    getbook(i)
```

```python
labels=[]
for i in url:
    S=input(" please enter book name ")
    labels.append(S)
def label(lst): #create Labels for every book 
    l1=labels[0]
    l2=labels[1]
    lst1 = [l1 for i in range(200)]
    lst2 = [l2 for i in range(200)]
    return lst1+lst2
labl=label(labels)
```

```python
def clean(holdb):    #Cleaning the data on books
    for i in range(len(holdb)):
        holdb[i]=holdb[i].lower()
        holdb[i]=re.sub("\n"," \n ",holdb[i])
        holdb[i]=holdb[i].replace(';',' ; ')
        holdb[i]=holdb[i].replace(',',' , ')
        holdb[i]=holdb[i].replace('.',' . ')
        holdb[i]=holdb[i].replace(';',' ; ')
        holdb[i]=holdb[i].replace('(',' ( ')
        holdb[i]=holdb[i].replace(':',' : ')
        holdb[i]=holdb[i].replace(':',' : ')
        holdb[i]=holdb[i].replace('?',' ? ')
        holdb[i]=holdb[i].replace('-',' - ')
        holdb[i] = holdb[i].split(' ')
        holdb[i] = ' '.join(filter(str.isalnum, holdb[i]))
        holdb[i] = re.sub("\d+", "", holdb[i])     
    for i in range(len(holdb)):
        holdb[i]=holdb[i].split(" ")
        while('' in holdb[i]) :  #remove empty elements 
            holdb[i].remove('')
clean(holdb)
```

```python
chunk_size=100
def chunkIt(lst): #split the book into equally batches 
    A=[]
    for i in range(len(lst)):
        chunked_list = list()
        for a in range(0, len(lst[i]), chunk_size):
            t=""
            for word in lst[i][a:a+chunk_size]:
                t+=word+" "
            chunked_list.append(t)
        A.append(chunked_list)
    return A
part=chunkIt(holdb) 
```

```python
Nlist=[]
def shuffle(t): #shuffling each book's partitions seperately 
    N=[]
    for i in range(len(t)):
        for a in range(0,len(t[i]),200):
            Nlist=rm.sample(t[i],200)
        N.append(Nlist)
    return N
X=shuffle(part)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None) #view options
f=pd.DataFrame() #create dataframe
f['Text']=X[0]+X[1] #put samples from book1,book2 in same column under the title Text
f['label']=labl #passing each partition to its matching label 
```
